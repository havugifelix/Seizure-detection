```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import cv2
import pickle
import os
from tqdm import tqdm
import mne
```

```python
bcg_path="/media/kashraf/Elements/seizure_detection/preprocessed/v1.5.2/EEG+Annoatations/eval/Background/"
seizure_path="/media/kashraf/Elements/seizure_detection/preprocessed/v1.5.2/EEG+Annoatations/eval/Seizures/"
# edf_path="/media/kashraf/Elements/seizure_detection/datasets/edf_resampled/train/01_tcp_ar/055/00005533/s003_2012_03_14
```

### Extract Sessions paths

```python
path_1="/media/kashraf/Elements/seizure_detection/datasets/edf/eval/01_tcp_ar/"
# path_2="/media/kashraf/Elements/seizure_detection/datasets/edf/dev/02_tcp_le/"
path_3="/media/kashraf/Elements/seizure_detection/datasets/edf/eval/03_tcp_ar_a/"
patients=[]
def session_fnames(path_1):
    patients=[]
    all_sess=[]
    subdir1=os.listdir(path_1)
    for fname in subdir1:
        subdir2= path_1+fname
        for patient in os.listdir(subdir2):
            sub_patient=path_1+fname+"/"+patient
#             patients.append(sub_patient)
            for sess in os.listdir(sub_patient):
                sess_path= path_1+fname+"/"+patient+"/"+sess
                all_sess.append(sess_path)
                
    return all_sess
```

```python
all_sess_paths=np.concatenate([session_fnames(path_1),session_fnames(path_3)])
```

### Save edf files and their annotations as pickle file

```python
not_resembled=[]
def get_annotation(session_path):
#     print(session_path)
    edf_path= session_path[:session_path.index("edf")]+"edf_resampled"+session_path[session_path.index("edf")+3:]
    fnames=[]
    annotations=[]
    files=os.listdir(session_path+"/")
#     print(len(files))
    for i in range(len(files)):
        if "edf" in files[i]:
            fnames.append(files[i][:-4])
    for file in fnames:
        ## Get annotations for each edf file in a session
        with open(session_path+"/"+file+".tse") as f:
            lines = f.readlines()
            annots=[k.split(" ")[:-1] for k in lines[2:]]
            Annots=np.array(annots).T
            onset=np.array(Annots[0],dtype=float)
            duration= np.array(Annots[1],dtype=float)-onset
            description=Annots[2]
            annotations=mne.Annotations(onset,duration,description=description)
            
        ## Read each each edf file with mne
        file_path=edf_path+"/"+file+".edf"
        if os.path.exists(file_path):
            data= mne.io.read_raw_edf(file_path)
            data.set_annotations(annotations)
        else:
            data=mne.io.read_raw_edf(session_path+"/"+file+".edf")
            data.set_annotations(annotations)
#             not_resembled.append(session_path+"/"+file+".edf")
            
#         data_annot={"edf":data,
#                      "annot":annots}
        
        ### Get seizure type
        count=0
        seizure=""
        for an in annots:
            if an[-1]!="bckg":
                count+=1
                seizure=seizure+an[-1]
        if count==0:
            f = open(bcg_path+file+"_bcgd.pkl","wb")
            pickle.dump(data,f)
            f.close()
#             pickle.dump(data_annot, bcg_path+file+"/_bcgd.pkl")
        else:
            
            f = open(seizure_path+file+"_"+seizure+".pkl","wb")
#             f = open(seizure_path+file+"_"+annots[count][-1]+".pkl","wb")
#             f = open(seizure_path+file+"_"+seizure_type+".pkl","wb")
            pickle.dump(data,f)
            f.close()
#             pickle.dump(data_annot, bcg_path+file+"/_"+annots[count][-1]+".pkl")
            
#         print(annots[count][-1])
#         annotations.append(data_annot)
        
for sess in tqdm(all_sess_paths):
    get_annotation(sess)
```
